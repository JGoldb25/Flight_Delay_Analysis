```python
#Import the libraries for the analysis
import pandas as pd

#Read in the data file as a Pandas data frame
df = pd.read_csv('Flight_Predictions.csv')

#Look at first several rows of data set
df.head()
```

```python
#View data shape (rows and columns) as well as columnar summary metrics
df.shape
df.describe()
```

```python
#Next, clean the data
#First, check for missing values
df.isnull().values.any() #There are missing values
df.isnull().values.sum() #There are 726,608 missing values

#Look at missing values by column
df.isnull().sum() #The last column 'Unnamed: 25' appears to have the majority of missing values

#Non-null values in last column
df['Unnamed: 25'] #This column isn't part of the flight data, it's due to the csv upload and can be removed

#Remove last column
df = df.drop('Unnamed: 25', axis=1)
```

```python
#View new data set
df.isnull().sum()

#Show rows which are cancelled or diverted flights
#df.loc[df['CANCELLED'] == 1]) |
#df.loc[df['DIVERTED'] == 1])
df_check = df.loc[df['CANCELLED'] == 1]) | df.loc[df['DIVERTED'] == 1])
#df.loc[df['ARR_DEL15'] == 'NaN'])


#Select 'Feature' columns of interest for analysis
df = df[["MONTH", "DAY_OF_MONTH", "DAY_OF_WEEK", "ORIGIN", "DEST", "CRS_DEP_TIME", "ARR_DEL15"]]

#Check for Null values by column again
df.isnull().sum()

#Show first five columns with missing values
df[df.isnull().values.any(axis=1)].head()

#Check to see the values changed for ARR_DEL15 (i.e row 54, 113, 179)
#These flights with missing arrival times are canceled or diverted flights
#Since these could be considered "late," mark them as ARR_DEL15 = 1
df = df.fillna({'ARR_DEL15': 1})
df.iloc[53:55]

#Additional Row Testing
#df.iloc[112:114]
#df.iloc[178:180]
```

```python
#Bin Departure Times
import math

for index, row in df.iterrows():
    df.loc[index, 'CRS_DEP_TIME'] = math.floor(row['CRS_DEP_TIME'] / 100)

df.head(10)

#Create Indicator columns for ORIGIN and DEST
#These variables represent categorical machine learning data, which must be quantized into binary integers (1 or 0)
#for the ML model
#Remove the previous columns for ORIGIN and DEST
df = pd.get_dummies(df, columns = ['ORIGIN','DEST'])
df.head()
```

```python
#View correlations in data
import matplotlib.pyplot as plt
import seaborn as sns

sns.heatmap(df, xticklabels=True, yticklabels=True)
plt.style.use('seaborn-colorblind')
plt.figure(figsize=(20,20))
sns.heatmap(df.corr(),
            linewidths = 0.25,
            square= True,
            cmap = 'seismic',
            linecolor = 'black',
            annot= True)


#df.describe()
#Rename certain columns
#print(df.rename(columns={'A': 'a', 'C': 'c'}))
```

```python
#Split data into training and testing
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(df.drop('ARR_DEL15', axis=1), df['ARR_DEL15'], test_size=0.25, random_state=42)

#Display X training data shape
train_x.shape

#Display X testing data shape
test_x.shape

#Display Y training data shape
train_y.shape

#Display Y testing data shape
test_y.shape

#Test Random Forest Model
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(random_state=13)
model.fit(train_x, train_y)

#Calculate Mean Accuracy of Model
predicted = model.predict(test_x)
model.score(test_x, test_y)


#Find Prediction Probabilities
from sklearn.metrics import roc_auc_score
probabilities = model.predict_proba(test_x)

#Find Roc Auc score
roc_auc_score(test_y, probabilities[:, 1])

#Create confusion matrix
from sklearn.metrics import confusion_matrix
confusion_matrix(test_y, predicted)

#Model Precision
from sklearn.metrics import precision_score

train_predictions = model.predict(train_x)
precision_score(train_y, train_predictions)

#Recall Score
from sklearn.metrics import recall_score
recall_score(train_y, train_predictions)


#Show output
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()

#Roc curve
from sklearn.metrics import roc_curve

fpr, tpr, _ = roc_curve(test_y, probabilities[:, 1])
plt.plot(fpr, tpr)
plt.plot([0, 1], [0, 1], color='grey', lw=1, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')


def predict_delay(departure_date_time, origin, destination):
    from datetime import datetime

    try:
        departure_date_time_parsed = datetime.strptime(departure_date_time, '%d/%m/%Y %H:%M:%S')
    except ValueError as e:
        return 'Error parsing date/time - {}'.format(e)

    month = departure_date_time_parsed.month
    day = departure_date_time_parsed.day
    day_of_week = departure_date_time_parsed.isoweekday()
    hour = departure_date_time_parsed.hour

    origin = origin.upper()
    destination = destination.upper()

    input = [{'MONTH': month,
              'DAY': day,
              'DAY_OF_WEEK': day_of_week,
              'CRS_DEP_TIME': hour,
              'ORIGIN_ATL': 1 if origin == 'ATL' else 0,
              'ORIGIN_DTW': 1 if origin == 'DTW' else 0,
              'ORIGIN_JFK': 1 if origin == 'JFK' else 0,
              'ORIGIN_MSP': 1 if origin == 'MSP' else 0,
              'ORIGIN_SEA': 1 if origin == 'SEA' else 0,
              'DEST_ATL': 1 if destination == 'ATL' else 0,
              'DEST_DTW': 1 if destination == 'DTW' else 0,
              'DEST_JFK': 1 if destination == 'JFK' else 0,
              'DEST_MSP': 1 if destination == 'MSP' else 0,
              'DEST_SEA': 1 if destination == 'SEA' else 0 }]

    return model.predict_proba(pd.DataFrame(input))[0][0]



predict_delay('1/10/2018 21:45:00', 'JFK', 'ATL')


predict_delay('2/10/2018 21:45:00', 'JFK', 'ATL')


predict_delay('2/10/2018 10:00:00', 'ATL', 'SEA')



import numpy as np

labels = ('Oct 1', 'Oct 2', 'Oct 3', 'Oct 4', 'Oct 5', 'Oct 6', 'Oct 7')
values = (predict_delay('1/10/2018 21:45:00', 'JFK', 'ATL'),
          predict_delay('2/10/2018 21:45:00', 'JFK', 'ATL'),
          predict_delay('3/10/2018 21:45:00', 'JFK', 'ATL'),
          predict_delay('4/10/2018 21:45:00', 'JFK', 'ATL'),
          predict_delay('5/10/2018 21:45:00', 'JFK', 'ATL'),
          predict_delay('6/10/2018 21:45:00', 'JFK', 'ATL'),
          predict_delay('7/10/2018 21:45:00', 'JFK', 'ATL'))
alabels = np.arange(len(labels))

plt.bar(alabels, values, align='center', alpha=0.5)
plt.xticks(alabels, labels)
plt.ylabel('Probability of On-Time Arrival')
plt.ylim((0.0, 1.0))


```
